```python
import matplotlib.pyplot as plt
 
# EX1
def ex1():
    x1 = [-5,5,5,-5,-5]
    y1 = [5,5,-5,-5,5]
    
    plt.plot(x1,y1)
    
    # naming the x axis
    plt.xlabel('x')
    # naming the y axis
    plt.ylabel('f(x)')
    # giving a title to my graph
    plt.title('Quadrado com 5 unidades de lado')
     
     
    # function to show the plot
    plt.show()

# Ex2
def ex2():
    nq=int(input("Introduza o n√∫mero de quadrados que pretende desenhar: "))
    
    qua=(5/nq)
    lQuadMax = 5
    x2= []
    y2= []
    
    for j in range (0,nq):
        lQua = lQuadMax - qua*j
        
        x2.append(-lQua)
        x2.append(lQua)
        x2.append(lQua)
        x2.append(-lQua)
        x2.append(-lQua)
        
        y2.append(lQua)
        y2.append(lQua)
        y2.append(-lQua)
        y2.append(-lQua)
        y2.append(lQua)
    
    plt.plot(x2,y2)
    
    # naming the x axis
    plt.xlabel('x')
    # naming the y axis
    plt.ylabel('f(x)')
    # giving a title to my graph
    plt.title('Quadrado concentricos')
     
     
    # function to show the plot
    plt.show()

ex1()
ex2()
```
