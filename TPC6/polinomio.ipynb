```python
#polinomio.y:módulo com as funcionalidades sobre polinómios
#2021-11-15
#
#Exemplos:
#p1 = x^2+2
#p2 = 7x^3+4x-21
#p3 = -x^3+2x^2-5
#modelo = lista de tuplos
#cada tuplo composto por coeficiente e expoente
#
import matplotlib.pyplot as plt

p1=[(1,2),(2,0)]
p2=[(7,3),(4,1),(-21,0)]
p3=[(-1,1),(2,2),(-5,0)]

ListaPol=[p1,p2,p3]

def criarTermo(coef,exp):
    return (coef,exp)

def criarPolinomio():
    return []

def inserirTermo(p,t):
    return p.append(t)

def verTermo(t):
    coef,exp=t
    if exp==0:
        return str(coef)
    elif exp==1:
        return (str(coef)+"x")
    elif coef==1:
        return ("x^"+str(exp))
    else:
        return(str(coef) + 'x^'+ str(exp))

def verPolinomio(p):
    res=""
    for t in p:
        if res=="":
            res=verTermo(t)
        else:
            res=res+ " + " + verTermo(t)
    
    return res

def verListaPol(lista):
    resL="id :: polinómio\n"
    i=1
    for p in ListaPol:
            resL=resL + "p" + str(i) + " :: " + verPolinomio(p) + "\n"
            i=i+1

    return resL


def calcPol(p,x):
    r=0
    for (coef,exp) in p:
        r=r+coef*(x**exp)
    
    return r

def tabela(p,nlinhas):
    for i in range (1,(nlinhas+1)):
        print(i, calcPol(p,i))

def chaveOrd(t):
    _,exp=t
    return exp

def ordenarPol(p):
    p.sort(reverse=True,key=chaveOrd)

def simplificarPol(p):
    ordenarPol(p)
    resLista=[]
    
    for i in range (1,len(p)):
        c,e=p[i]
        c0,e0=p[i-1]
        if e0==e:
            inserirTermo(resLista, (c+c0,e))
        
        else:
            if e0>e:
                if i==1:
                    inserirTermo(resLista, (c0,e0))
            
            else:
                inserirTermo(resLista, (c,e))
    
    return resLista


def derivarPol(p):
    r=[]
    for (coef,exp) in p:
        if exp>0:
            r.append((coef*exp,exp-1))
    
    return r
```
