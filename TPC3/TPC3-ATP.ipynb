```python
import random

def menu():
    print ('''
           Gestão de listas de inteiros
             1-Introduzir a lista à mão
             2-Gerar lista com valores aleatórios
             3-Maior elemento da lista
             4-Ordenar de modo crescente a lista
             0-Sair''')
    
def criarListaM():
    n=int(input("Introduza o número de elementos da lista: "))
    L=[]
    i=0

    while i<n:
        N=int(input("introduza o elemento %s: "%(i+1)))   
        L.append(N) 
        i+=1
    
    return L

def criarListaA():
    n=int(input("Introduza o número de elementos da lista: "))
    L2=[]
    i=0
    
    while i<n:
        N=random.randint(0,100)   
        L2.append(N) 
        i+=1
    
    return L2

def maximoLista(l):
    n=len(l)
    maximo=l[0]
    i=1
    while i<n:
        if maximo<l[i]:
            maximo=l[i]
        
        i+=1
    return maximo

def trocasDiretas(l2):
    t=True
    while t==True:
        i=0
        trocas=0
        while i<(len(l2)-1):
            if l2[i]>l2[i+1]:
                temp=l2[i]
                l2[i]=l2[i+1]
                l2[i+1]=temp
                trocas=1
            i+=1
        
        if trocas!=1:
            t=False
        else:
            t=True
    
    return (l2)
        
op="a"

while op!=0:  
    menu() 
    op=int(input("Introduza a opção pretendida: "))          
    if op==1:
        lis=criarListaM()
        print(lis)
    
    elif op==2:
        lis=criarListaA()
        print(lis)
    
    elif op==0:
        print("Até à próxima")
    
    elif op==3:
        print(maximoLista(lis))
    
    elif op==4:
        print(trocasDiretas(lis))
    
    else:
        print("Opção inválida")
```
